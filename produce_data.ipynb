```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime
from sklearn.utils import resample
```

```python
# Initialize Faker
fake = Faker()

# Load the original dataset
df = pd.read_csv('telco_churn.csv')

display(df)
```

```python
def generate_synthetic_data(original_df, num_samples=7041):
    """
    Generate synthetic data that maintains similar statistical properties to the original dataset
    """
    synthetic_data = []
    
    # Get value counts for categorical columns to maintain distributions
    churn_dist = original_df['Churn'].value_counts(normalize=True)
    contract_dist = original_df['Contract'].value_counts(normalize=True)
    internet_type_dist = original_df['Internet Type'].value_counts(normalize=True)
    payment_dist = original_df['Payment Method'].value_counts(normalize=True)
    gender_dist = original_df['Gender'].value_counts(normalize=True)
    
    # Get numerical column statistics
    age_stats = original_df['Age'].describe()
    monthly_gb_stats = original_df['Avg Monthly GB Download'].describe()
    monthly_charge_stats = original_df['Monthly Charge'].describe()
    tenure_stats = original_df['Tenure in Months'].describe()
    
    # Churn reasons and categories
    churn_reasons = original_df['Churn Reason'].dropna().unique()
    churn_categories = original_df['Churn Category'].dropna().unique()
    
    for _ in range(num_samples):
        # Generate customer demographics
        gender = np.random.choice(gender_dist.index, p=gender_dist.values)
        age = int(np.random.normal(age_stats['mean'], age_stats['std']))
        age = max(18, min(age, 81))  # Ensure reasonable age range
        married = random.choice([0, 1])
        dependents = random.choices([0, 1, 2, 3], weights=[0.65, 0.15, 0.15, 0.05])[0]
        
        # Generate location data
        city = fake.city()
        state = "California"
        state_abbr = "CA"  # Using state abbreviation for zip code generation
        try:
            zip_code = fake.zipcode_in_state(state_abbr)
        except:
            zip_code = fake.zipcode()  # Fallback if state abbreviation fails
            
        lat = round(np.random.uniform(32.55, 41.96), 6)
        lng = round(np.random.uniform(-124.30, -114.19), 6)
        lat_long = f"{lat}, {lng}"
        population = random.randint(11, 101215)
        
        # Generate service information
        contract = np.random.choice(contract_dist.index, p=contract_dist.values)
        phone_service = random.choice([0, 1])
        multiple_lines = random.choice([0, 1]) if phone_service else 0
        internet_service = random.choice([0, 1])
        
        internet_type = None
        if internet_service:
            internet_type = np.random.choice(internet_type_dist.index, p=internet_type_dist.values)
        
        # Generate usage data
        avg_monthly_gb = max(0, round(np.random.normal(monthly_gb_stats['mean'], monthly_gb_stats['std']), 2))
        avg_long_dist = round(np.random.uniform(0, 50), 2) if random.random() > 0.3 else 0
        monthly_charge = max(10, round(np.random.normal(monthly_charge_stats['mean'], monthly_charge_stats['std']), 2))
        tenure = max(1, int(np.random.normal(tenure_stats['mean'], tenure_stats['std'])))
        
        # Generate churn-related data
        churn = np.random.choice(churn_dist.index, p=churn_dist.values)
        churn_category = None
        churn_reason = None
        churn_score = random.randint(5, 96)
        
        if churn == 1:
            churn_category = random.choice(churn_categories)
            churn_reason = random.choice(churn_reasons)
            churn_score = random.randint(65, 96)  # Higher scores for churned customers
        else:
            churn_score = random.randint(5, 64)  # Lower scores for non-churned
            
        # Generate customer status
        status_options = ['Stayed', 'Joined', 'Churned']
        customer_status = 'Churned' if churn == 1 else random.choice(['Stayed', 'Joined'])
        
        # Generate other features
        paperless_billing = random.choice([0, 1])
        payment_method = np.random.choice(payment_dist.index, p=payment_dist.values)
        referred_friend = random.choice([0, 1])
        number_referrals = random.randint(0, 10) if referred_friend else 0
        satisfaction_score = random.randint(1, 5)
        senior_citizen = 1 if age >= 65 else 0
        partner = random.choice([0, 1])
        
        # Generate device and service features
        device_protection = random.choice([0, 1]) if internet_service else 0
        online_backup = random.choice([0, 1]) if internet_service else 0
        online_security = random.choice([0, 1]) if internet_service else 0
        premium_tech_support = random.choice([0, 1]) if internet_service else 0
        streaming_tv = random.choice([0, 1]) if internet_service else 0
        streaming_movies = random.choice([0, 1]) if internet_service else 0
        streaming_music = random.choice([0, 1]) if internet_service else 0
        unlimited_data = random.choice([0, 1]) if internet_service else 0
        
        # Generate financial metrics
        total_charges = round(monthly_charge * tenure * random.uniform(0.9, 1.1), 2)
        total_long_dist = round(avg_long_dist * tenure * random.uniform(0.9, 1.1), 2)
        total_extra_data = round(random.uniform(0, 100), 2) if random.random() > 0.8 else 0
        total_refunds = round(random.uniform(0, 50), 2) if random.random() > 0.9 else 0
        total_revenue = total_charges + total_long_dist + total_extra_data - total_refunds
        
        # Generate customer ID
        customer_id = f"{fake.random_number(digits=4)}-{fake.lexify(text='?????', letters='ABCDEFGHIJKLMNOPQRSTUVWXYZ')}"
        
        # Generate quarter (assuming Q3 based on original data)
        quarter = "Q4"
        
        # Generate CLTV (Customer Lifetime Value)
        cltv = random.randint(2003, 6500)
        
        # Generate offer
        offers = ['Offer A', 'Offer B', 'Offer C', 'Offer D', 'Offer E', None]
        offer = random.choice(offers)
        
        # Generate under 30 flag
        under_30 = 1 if age < 30 else 0
        
        # Create the synthetic record
        record = {
            'Age': age,
            'Avg Monthly GB Download': avg_monthly_gb,
            'Avg Monthly Long Distance Charges': avg_long_dist,
            'Churn Category': churn_category,
            'Churn Reason': churn_reason,
            'Churn Score': churn_score,
            'City': city,
            'CLTV': cltv,
            'Contract': contract,
            'Country': "United States",
            'Customer ID': customer_id,
            'Customer Status': customer_status,
            'Dependents': dependents,
            'Device Protection Plan': device_protection,
            'Gender': gender,
            'Internet Service': internet_service,
            'Internet Type': internet_type,
            'Lat Long': lat_long,
            'Latitude': lat,
            'Longitude': lng,
            'Married': married,
            'Monthly Charge': monthly_charge,
            'Multiple Lines': multiple_lines,
            'Number of Dependents': dependents,
            'Number of Referrals': number_referrals,
            'Offer': offer,
            'Online Backup': online_backup,
            'Online Security': online_security,
            'Paperless Billing': paperless_billing,
            'Partner': partner,
            'Payment Method': payment_method,
            'Phone Service': phone_service,
            'Population': population,
            'Premium Tech Support': premium_tech_support,
            'Quarter': quarter,
            'Referred a Friend': referred_friend,
            'Satisfaction Score': satisfaction_score,
            'Senior Citizen': senior_citizen,
            'State': state,
            'Streaming Movies': streaming_movies,
            'Streaming Music': streaming_music,
            'Streaming TV': streaming_tv,
            'Tenure in Months': tenure,
            'Total Charges': total_charges,
            'Total Extra Data Charges': total_extra_data,
            'Total Long Distance Charges': total_long_dist,
            'Total Refunds': total_refunds,
            'Total Revenue': total_revenue,
            'Under 30': under_30,
            'Unlimited Data': unlimited_data,
            'Zip Code': zip_code,
            'Churn': churn
        }
        
        synthetic_data.append(record)
    
    return pd.DataFrame(synthetic_data)
```

```python
# Generate synthetic data (1000 samples)
synthetic_df = generate_synthetic_data(df, num_samples=7041)

# Save to CSV
synthetic_df.to_csv('synthetic_telco_churn.csv', index=False)

print("Synthetic data generated and saved to 'synthetic_telco_churn.csv'")
```
